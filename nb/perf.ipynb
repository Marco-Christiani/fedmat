````python
import plotly.express as px
````

````python
from pathlib import Path
import json

output_dir = Path("../outputs/vit_cifar10")
models = {
    "Pretrained": output_dir.joinpath("metrics-pretrained-2025-11-15T04:45:04.758564.json"),
    "Raw": output_dir.joinpath("metrics-raw.json")
}
````

````python
for name, path in models.items():
    metrics = json.loads(path.read_text())
    metrics_train = metrics["train"]
    px.line(
        x=metrics_train["global_step"],
        y=metrics_train["loss"]
    ).update_layout(
        title=f"{name} ViT CIFAR Training",
        xaxis_title="Global step",
        yaxis_title="Loss",
    ).show()
````

# Confusion mat

````python
import torch

conf = torch.load(output_dir/"confusion_matrix-2025-11-15T05:03:15.777355.pt").numpy()

fig = px.imshow(
    conf,
    text_auto=True,
    color_continuous_scale="Viridis",
    labels={"x": "Predicted", "y": "True"},
)
fig.update_layout(
    width=500, height=500,
    title="Confusion Matrix"
)
fig.show()
````
